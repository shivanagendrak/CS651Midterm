```python
#Question 2
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/shiva/OneDrive/Desktop/MidtermData.csv")
data["text"] = data["Job Title"] + " " + data["Job Description"]
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(data["text"])
feature_names = vectorizer.get_feature_names()
tfidf_values = tfidf[0].toarray()[0]
tfidf_dict = dict(zip(feature_names, tfidf_values))
wordcloud = WordCloud(width=800, height=800, background_color="black")
wordcloud.generate_from_frequencies(tfidf_dict)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
#Question 4
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/shiva/OneDrive/Desktop/MidtermData.csv")
polarities = []
for job_description in data["Job Description"]:
    blob = TextBlob(job_description)
    polarities.append(blob.sentiment.polarity)
plt.hist(polarities, bins=20)
plt.title("Sentiment Analysis of Job Descriptions")
plt.xlabel("Sentiment Polarity")
plt.ylabel("Frequency")
plt.show()
```
